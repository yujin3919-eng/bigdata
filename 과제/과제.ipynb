```python
# 1

class Product:
    """ìƒí’ˆ ì •ë³´ë¥¼ ê´€ë¦¬í•˜ëŠ” í´ë˜ìŠ¤"""
    
    def __init__(self, product_id, name, price, weight, stock):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.weight = weight
        self.stock = stock
    
    def get_info(self):
        """ìƒí’ˆ ì •ë³´ë¥¼ ë”•ì…”ë„ˆë¦¬ë¡œ ë°˜í™˜"""
        return {
            'product_id': self.product_id,
            'name': self.name,
            'price': self.price,
            'weight': self.weight,
            'stock': self.stock
        }
    
    def update_stock(self, quantity):
        """
        ì¬ê³  ìˆ˜ëŸ‰ ì—…ë°ì´íŠ¸
        ì¬ê³ ê°€ ë¶€ì¡±í•˜ë©´ False ë°˜í™˜, ì„±ê³µí•˜ë©´ True ë°˜í™˜
        """
        if self.stock >= quantity:
            self.stock -= quantity
            return True
        else:
            return False
    
    def apply_discount(self, discount_rate):
        """í• ì¸ìœ¨ì„ ì ìš©í•œ ê°€ê²© ë°˜í™˜ (ì›ë³¸ ê°€ê²©ì€ ìœ ì§€)"""
        discounted_price = int(self.price * (1 - discount_rate))
        return discounted_price


# 2 

class ShoppingCart:
    """ì¥ë°”êµ¬ë‹ˆë¥¼ ê´€ë¦¬í•˜ëŠ” í´ë˜ìŠ¤"""
    
    def __init__(self):
        self.items = []
    
    def add_product(self, product, quantity):
        if product.update_stock(quantity):
            self.items.append({"product": product, "quantity": quantity})
            print(f"{product.name} {quantity}ê°œê°€ ì¥ë°”êµ¬ë‹ˆì— ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤.")
        else:
            print("ì¬ê³ ê°€ ë¶€ì¡±í•©ë‹ˆë‹¤.")
    
    def remove_product(self, product_id):
        for item in self.items:
            if item["product"].product_id == product_id:
                self.items.remove(item)
                print("í•´ë‹¹ ìƒí’ˆì´ ì¥ë°”êµ¬ë‹ˆì—ì„œ ì œê±°ë˜ì—ˆìŠµë‹ˆë‹¤.")
                return
        print("í•´ë‹¹ ìƒí’ˆì´ ì¥ë°”êµ¬ë‹ˆì— ì—†ìŠµë‹ˆë‹¤.")
    
    def get_total_price(self):
        total = 0
        for item in self.items:
            total += item["product"].price * item["quantity"]
        return total
    
    def get_total_weight(self):
        total_weight = 0
        for item in self.items:
            total_weight += item["product"].weight * item["quantity"]
        return total_weight
    
    def show_cart(self):
        print("\n===== ì¥ë°”êµ¬ë‹ˆ =====")
        for item in self.items:
            product = item["product"]
            quantity = item["quantity"]
            print(f"ìƒí’ˆëª… ë° ìˆ˜ëŸ‰: {product.name}[{quantity}]")
            print(f"ê°€ê²© : {product.price*quantity}ì›")
        print("==================\n")


# 3
class Order:
    """ì£¼ë¬¸ ì •ë³´ë¥¼ ê´€ë¦¬í•˜ëŠ” í´ë˜ìŠ¤"""
    
    SHIPPING_FREE_THRESHOLD = 50000
    SHIPPING_FEE = 3000     

    def __init__(self, order_id, cart, customer_name):
        self.order_id = order_id
        self.cart = cart
        self.customer_name = customer_name
    
    def calculate_shipping_fee(self):
        """ë°°ì†¡ë¹„ ê³„ì‚° (5ë§Œì› ì´ìƒ ë¬´ë£Œë°°ì†¡)"""
        total_price = self.cart.get_total_price()
        if total_price >= self.SHIPPING_FREE_THRESHOLD:
            return 0
        return self.SHIPPING_FEE
    
    def calculate_final_price(self):
        """ìµœì¢… ê²°ì œ ê¸ˆì•¡ ê³„ì‚° (ìƒí’ˆ ê¸ˆì•¡ + ë°°ì†¡ë¹„)"""
        return self.cart.get_total_price() + self.calculate_shipping_fee()
    
    def print_receipt(self):
        product_total = self.cart.get_total_price()
        shipping_fee = self.calculate_shipping_fee()
        final_price = self.calculate_final_price()

        print("\n===== ì£¼ë¬¸ì„œ =====")
        print(f"ì£¼ë¬¸ ID: {self.order_id}")
        print(f"ê³ ê°ëª…: {self.customer_name}")
        
        # show_cart() ëŒ€ì‹  ì§ì ‘ ì¶œë ¥
        print("\n[ì£¼ë¬¸ ìƒí’ˆ ëª©ë¡]")
        for item in self.cart.items:
            product = item["product"]
            quantity = item["quantity"]
            print(f"  - {product.name} x {quantity}ê°œ = {product.price * quantity}ì›")
        
        print(f"\nìƒí’ˆ ê¸ˆì•¡: {product_total}ğŸ’³")
        print(f"ë°°ì†¡ë¹„: {shipping_fee}ğŸ’³")
        print(f"ìµœì¢… ê²°ì œ ê¸ˆì•¡: {final_price}ğŸ’³")
        print("==================\n")




# ìƒí’ˆ ìƒì„±
laptop = Product("P001", "ë…¸íŠ¸ë¶", 1200000, 2.5, 10)
mouse = Product("P002", "ë¬´ì„ ë§ˆìš°ìŠ¤", 35000, 0.2, 50)
keyboard = Product("P003", "ê¸°ê³„ì‹í‚¤ë³´ë“œ", 89000, 1.0, 30)

# ì¥ë°”êµ¬ë‹ˆ ìƒì„± ë° ìƒí’ˆ ì¶”ê°€
print()
cart = ShoppingCart()
cart.add_product(laptop, 1)
cart.add_product(mouse, 2)
cart.add_product(keyboard, 1)

# ì¥ë°”êµ¬ë‹ˆ í™•ì¸
cart.show_cart()

# ì£¼ë¬¸ ìƒì„±
order = Order("ORD20260211001", cart, "í™ê¸¸ë™")

# ì£¼ë¬¸ì„œ ì¶œë ¥
order.print_receipt()
```
