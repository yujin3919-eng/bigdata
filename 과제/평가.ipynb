````python
# 문제 1 
# 1부터 50까지의 숫자 중 짝수만 출력하는 프로그램을 작성하시오.

a = list()
b = list()

for i in range(1, 51):
    if i % 2 == 0:
        a.append(i)
    else:
        b.append(i)

print(f"짝수: {a}")
````

````python
# 문제 2 
# 사용자로부터 숫자를 입력받아 해당 숫자의 구구단을 출력하는 프로그램을 작성하시오.

dan = int(input("몇 단?: "))

if dan < 2 or dan > 9:
    print("2~9단 내에서 선택하세요.")
else:
    for i in range(1, 10):
        print(f"{dan}x{i}={dan*i}")

````

````python
# 문제 3 
# 다음과 같은 패턴을 출력하는 프로그램을 작성하시오.

# **출력:**
# ```
# *
# **
# ***
# ****
# *****


def left_star(height):
    print(f"**출력:{height}**")
    print("```")

    for i in range(1, height+1):
        print("*" * i)


left_star(5)

````

````python
# 문제 4
# 두 개의 숫자를 입력받아 두 수 사이의 모든 정수의 합을 반환하는 함수 sum_between(a, b)를 작성하시오.
# (a와 b 포함)

def sum_between(a, b):
    if a > b:
        a, b = b, a
    return sum(range(a, b+1))

try:
    a = int(input("숫자1: "))
    b = int(input("숫자2: "))
    print(sum_between(a, b))
except ValueError:
    print("숫자만 입력하세요.")
````
