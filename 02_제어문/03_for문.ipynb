```python
# FOR문의 기본구조

# for 변수 in 리스트(또는 튜플, 문자열):
#     수행할 문장1
#     수행할 문장2

lis = [1, 2, 3]
```

```python
# 전형적인 for 문
test_list = ["one", "two", "three"]

for i in test_list:
    print(i)
```

```python
# 다양한 for 문의 사용
a = [(1, 2), (3, 4), (5, 6)]

for (a, b) in a:
    print(a + b)
```

```python
# for 문의 응용

# 5명의 시험 점수
scores = [90, 25, 67, 45, 80]
id = 0

for score in scores:
    id += 1   # id = id + 1
    if score >= 60:
        print(f"id: {id}, 합격입니다.")
    else:
        print(f"id: {id}, 불합격입니다.")
```

```python
# for문과 continue문
scores = [90, 25, 67, 45, 80]
id = 0

for score in scores:
    id += 1

    if score < 60:
        continue

print(f"id가 {id} 학생은 합격입니다.")
```

```python
# for문과 함께 자주 사용하는 range 함수

a = range(10)
print(list(a))

b = range(1, 10)
list(b)
```

```python
# 1부터 10까지 더하는 반복문

hap = 0

for i in range(1, 10+1):   # [1, 2, 3, ... 10]
    hap = hap + i
    print(hap)

print(hap)
```
