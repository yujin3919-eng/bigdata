```python
# 곱하기, 빼기, 나누기 기능 만들기
## mul() - 곱하기
## sub() - 빼기
## div() - 나누기

### 계산기 2개 만들어서 (cal1, cal2)
### cal1(10, 20)
### cal2(20, 5)
```

```python
class FourCal:
     def setdata(self, first, second):
         self.first = first
         self.second = second
     def add(self):
         result = self.first + self.second
         return result
     def mul(self):
         result = self.first * self.second
         return result
     def sub(self):
         result = self.first - self.second
         return result
     def div(self):
         result = self.first / self.second
         return result
     
cal1 = FourCal()
cal1.setdata(10, 20)
print(cal1.add())
print(cal1.mul())
print(cal1.sub())
print(cal1.div())

print("====")

cal2 = FourCal()
cal2.setdata(20, 5)
print(cal2.add())
print(cal2.mul())
print(cal2.sub())
print(cal2.div())

```
